```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.array([1, 2,4, 30])
y = np.array([3, 1, 4, 10])
plt.plot(x, y, color="g")
plt.title("line wala graph")
plt.xlabel("x ki values")
plt.ylabel("y ki values")
```

```python
# bar graph
name = ["ali", "ahmad", "zubair", "jhangir"]
marks = [20, 23, 12, 23]

plt.bar(name, marks)
plt.title("bar wala graph")
plt.show()
```

```python
# histogram
random_data = np.random.randn(1000)
plt.hist(random_data)
plt.show()
```

```python
# scatter plot
x = [1, 3, 5, 6, 9]
y = [4, 2, 1,5 , 6]
plt.scatter(x, y, color='red')
plt.title("scatter plot")
plt.show()
```

```python
subjects = ["physics", "chemistry", "biology", "computer"]
marks = [12, 20, 14, 14]
plt.pie(marks, labels= subjects, autopct='%1.1f%%')
plt.title("pie wala graph")
plt.show()
```
