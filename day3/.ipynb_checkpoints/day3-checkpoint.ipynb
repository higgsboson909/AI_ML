```python
# if you want to check the type of array
# for this a property is used

a = np.array([1])
b = np.array([1, 3, 5, 7])
c = np.array([[[2, 3],[2, 3], [3, 4]]])

print(a.ndim)
print(b.ndim)
print(c.ndim)
```

```python
# accessing an element through index
arr6 = np.array([1,2,3,4,5])
print(arr6[4])
```

```python
# slicing array -> 
arr7 = np.array([3,4,1,2,3,4,5])
print(arr7[2:4])
```

```python
#basic operations of numpy library
arr11 = np.array([10, 20, 30, 40])
print("mean", np.mean(arr11))
print("max", np.max(arr11))
print("min", np.min(arr11))
print("sum", np.sum(arr11))

```

```python
# common operations of pandas that every data scientist or AI engineer should know.

# head
# tail
# shape -> size of dataset
# column -> column features
# iloc -> indexing
# describe() -> stats
# filter -> setting data
# val_count() -> for counting
# sort_values() -> sort data
```
